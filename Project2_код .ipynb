```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df_18 = pd.read_csv('2018.csv')
df_19 = pd.read_csv('2019.csv')
```

Объединяем таблицы с данными. Вводим дополнительные столбцы обновления данных. 
Отбираем данные по штатам.
Сортируем по городам и групперуем по датам.

```python
# Список штатов.
states = ['OH', 'NV', 'FL', 'IN', 'AL', 'OR']
# Список городовю
city_to_select = ['Columbus', 
                  'Cincinnati', 
                  'Cleveland', 
                  'Las Vegas', 
                  'Miami',
                  'Indianapolis',
                  'Birmingham',
                  'Portland']  
# Вводим маркеры для датафреймов 2018 и 2019г.
year = ['2018', '2019']
df_18['group'] = year[0]
df_19['group'] = year[1]
# Названия кафе приводим к одному регистру и убираем лишние символы
df_main = pd.concat([df_18,df_19]).reset_index(drop=1)
```

```python
df_main.head()
```

```python
df_main['name'] = df_main['name'].apply(lambda x: x.lower().replace("'", "").replace(' ', '').replace('¬æ', ''))
# Отфильтровываем данные по штатам и городам.
df_filtred = df_main[(df_main['province'].isin(states)) 
                     & (df_main['city'].isin(city_to_select))].reset_index(drop=1)
df_filtred.loc[371, 'name'] = df_filtred['name'][371].split('-')[0]
# Групируем по названию кафе и городам. Получаем данные по числу кафе каждой компании в каждом городе.
df_grouped = (df_filtred
             .groupby(['group','city','name'])
             .agg(cnt=('name','count'))
             .sort_values(by=['city','cnt'], ascending=False)
             .reset_index()).copy()
```

Расчитываем среднее количество заведений на 10000 человек.

```python
# Добавляем к датафрейму данные населения городов (данные взяты с ресурса: www.opendatanetwork.com)
pop_18 = [867628, 300357, 387398, 626637, 451214, 857637, 212021, 639387]
pop_19 = [880301, 296778, 381663, 634307, 454851, 864276, 209857, 649431]
# Добавляем к датафрейму данные туристического потока городов
turism = [41900000, 26000000, 19600000, 49500000, 24200000, 29200000, 3900000, 12500000]
df_pop = pd.DataFrame(data={'city':city_to_select,
                             '2018':pop_18,
                             '2019':pop_19,
                             'tourism': turism})

# Групперуем по городам и датам. Получаем данные по общему числу кафе в каждом городе.
df_quant = (df_grouped.groupby(['group', 'city'])
                    .agg(quantity=('cnt','sum'))
                    .sort_values(by=['city','group','quantity'], ascending=True)
                    .reset_index())

# Добавляем колонки с населением и количеством кафе на 10К человек для каждого города.
for i in range(len(df_quant)):
    for j in range(len(df_pop)):
        if df_quant['city'][i] == df_pop['city'][j]:
            for q in range(len(df_pop.columns)):
                if df_quant['group'][i] == df_pop.columns[q]:
                    df_quant.loc[i,'population'] = int(df_pop[df_pop.columns[q]][j])
                    df_quant.loc[i,'cafe on 10K persons'] = round((df_quant['quantity'][i] *10000 / df_pop[df_pop.columns[q]][j] ), 2)
                    df_quant.loc[i,'10K person on cafe'] = round(df_pop[df_pop.columns[q]][j] / (df_quant['quantity'][i] *10000), 2)
                elif (df_quant['group'][i] == '2019') & (df_pop.columns[q] == 'tourism'):
                    df_quant.loc[i, '10K turist on cafe'] = round(df_pop[df_pop.columns[q]][j] / (df_quant['quantity'][i] *10000), 2)
df_quant['total'] = df_quant[df_quant['group'] == '2019']['10K person on cafe'] + df_quant[df_quant['group'] == '2019']['10K turist on cafe']
                
```

Посмотрим изменение количества кафе за 2018 и 2019.

```python
plt.rcParams['figure.figsize'] = (10,10) #Размер графиков
sns.set_theme(style='whitegrid', palette='hls') # Цветовая настройка

sns.barplot(df_quant.sort_values(['quantity'], ascending=[False], inplace=False), 
            x='quantity',
            y='city', 
            hue='group',
            errorbar=None)

plt.title('Количество ресторанов фастфуд.')
plt.ylabel(None)
plt.xlabel('Количество')

plt.show()
```

Из диаграммы видно, что в 6 городах количество кафе увеличилось, а в 2 уменьшилось.

Посмотрим на компании, которые были открыты в 2018, но закрылись в 2019.
и наоборот на вновь открытые. В каждом городе.

```python
# Количество компаний в городах в 18 и 19гг.
quant_in_city_18 = []
quant_in_city_19 = []
# Количество компании, которые были на рынке в 2018г, но не было 2019г.
close_in_city = []
# Количество компании, которые остались 2019г.
old_in_city = []
# Количество компании, которые появились на рынке 2019г.
open_new_in_city = []
# Количество ресторанов, которые были открыты или закрыты в городах.
open_in_city = []
# Список компаний которые закрылись.
close_in_city_list = []
# Список компаний которые открылись.
open_in_city_list = []
# Список компаний которые сохранили присутствие на рынке.
old_in_city_list = []

for i in city_to_select:
    quant_in_city_18.append(len(set(df_filtred[(df_filtred['group'] == '2018') & (df_filtred['city'] == i)]['name'])))

for i in city_to_select:
    quant_in_city_19.append(len(set(df_filtred[(df_filtred['group'] == '2019') & (df_filtred['city'] == i)]['name'])))

for i in city_to_select:
    close_in_city_list.append(list((set(df_filtred[(df_filtred['group'] == '2018') & (df_filtred['city'] == i)]['name'])
                              .difference(set(df_filtred[(df_filtred['group'] == '2019') & (df_filtred['city'] == i)]['name'])))))
    
for i in city_to_select:
    open_in_city_list.append(list((set(df_filtred[(df_filtred['group'] == '2019') & (df_filtred['city'] == i)]['name'])
                                 .difference(set(df_filtred[(df_filtred['group'] == '2018') & (df_filtred['city'] == i)]['name'])))))
    
for i in city_to_select:
    old_in_city_list.append(list((set(df_filtred[(df_filtred['group'] == '2018') & (df_filtred['city'] == i)]['name'])
                                 .intersection(set(df_filtred[(df_filtred['group'] == '2019') & (df_filtred['city'] == i)]['name'])))))
    
for i in range(len(close_in_city_list)):
    close_in_city.append(len(close_in_city_list[i]))

for i in range(len(close_in_city_list)):
    open_new_in_city.append(len(open_in_city_list[i]))

for i in range(len(close_in_city_list)):
    old_in_city.append(len(old_in_city_list[i]))
    
for i in city_to_select:
    open_in_city.append(int(df_quant[(df_quant['group'] == '2019') & (df_quant['city'] == i)]['quantity'])
                         - int(df_quant[(df_quant['group'] == '2018') & (df_quant['city'] == i)]['quantity']))
    
df_info = pd.DataFrame(data={'city': city_to_select,
                             '2018': quant_in_city_18,
                             '2019': quant_in_city_19,
                             'close': close_in_city,
                             'old': old_in_city,
                             'open_new': open_new_in_city,
                             'open': open_in_city})

df_info
```

Из данной таблицы видно, что в 2019г. на рынок вышло порядка 20 новых компаний в каждом городе, даже не смотря на то, что в Лас-вегасе и Майями произошло сокращение рынка.


Выведем данные по трем категорям: 
1 - df_close - рестораны закрылись, т.е. которые были в датафрейме 2018г. но отсутствуют в 2019г.
2 - df_old - рестораны старые, которые были и в 2018г и 2019г.
3 - df_new - рестораны новые, которых не было в 2018г., но они есть в 2019г.

```python
df_сomp = pd.DataFrame(data={'city': city_to_select,
                             'close': close_in_city_list,
                             'old': old_in_city_list,
                             'open': open_in_city_list})

close = []
open = []
old = []

for i in city_to_select:
    close.append(df_filtred[(df_filtred['city'] == i) 
                            & (df_filtred['name'].isin(list(df_сomp[df_сomp['city'] == i]['close'])[0]))])


for i in city_to_select:
    open.append(df_filtred[(df_filtred['city'] == i) 
                            & (df_filtred['name'].isin(list(df_сomp[df_сomp['city'] == i]['open'])[0]))])


for i in city_to_select:
    old.append(df_filtred[(df_filtred['city'] == i) 
                            & (df_filtred['name'].isin(list(df_сomp[df_сomp['city'] == i]['old'])[0]))])

df_closed = pd.concat(close)
df_new = pd.concat(open)
df_old = pd.concat(old)

df_closed.to_csv('closed.csv')
df_new.to_csv('new.csv')
df_old.to_csv('old.csv')
```

Посмотрим вероятность успешного развития бизнеса в зависимости от количества ресторанов.

Определим её как отношение старых ресторанов к сумме старых и закрытых.

```python
df_old_grouped = df_old.groupby(['city','name']).agg(cnt=('name','count'))
df_close_grouped = df_closed.groupby(['city','name']).agg(cnt=('name','count'))


posible = [0]
iter = [0]
for i in range(1,11):
    iter.append(i)
    posible.append(round((len(df_old_grouped[df_old_grouped['cnt'] == i]) 
                                    
                       / (len(df_old_grouped[df_old_grouped['cnt'] == i]) 
                                    
                     + len(df_close_grouped[df_close_grouped['cnt'] == i]))), 2))

posible.pop(0)
iter.pop(0)

print(posible)

sns.lineplot(x=iter, y=posible)

plt.title('Вероятность успеха в зависимости от количества ресторанов')
plt.xlabel('Количество ресторанов')

plt.ylabel('P')
plt.show()
```

Из графика видно, что рестораны, которые были в единичном экземпляре не удержались на рынке.
